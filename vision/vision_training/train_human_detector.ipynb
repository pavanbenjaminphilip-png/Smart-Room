# 🧠 Smart Room: High-Precision Human Detector (YOLOv8)

Following the goal to **maximize human detection capability**, this notebook is optimized for training a model that recognizes humans in any pose (sitting, standing, lying down) while ignoring environmental clutter.

### Key Improvements for Human Accuracy:
1. **Class Filtering**: Force the model to focus 100% on the **Person** class (Class 0).
2. **Pose Diversity**: We use the YOLOv8 pre-trained weights which are already expert at human shapes.
3. **TFLite Precision**: Exporting with high-precision TFLite parameters for the RPi 5.

```python
# 1. Install Ultralytics (YOLOv8)
!pip install ultralytics opencv-python tensorflow
```

```python
# 2. Initialize Model
from ultralytics import YOLO

# We use 'yolov8n' (Nano) for 30 FPS. If you want even MORE accuracy, swap to 'yolov8s' (Small).
model = YOLO('yolov8n.pt')
```

### Training Strategy
For maximum capability, we train for more epochs (25+) and ensure the model prioritizes the human class. 

**Note**: `coco8.yaml` is for a quick test. For full deployment, the system will automatically download the required weights.

```python
# 3. Train on Humans (Classes=0 is the Person class in COCO)
results = model.train(
    data='coco8.yaml', 
    epochs=25, 
    imgsz=640, 
    classes=[0],  # CRITICAL: Focus ONLY on humans
    plots=True
)
```

### Exporting for Raspberry Pi 5
We use `format='tflite'` with `int8=True` for maximum speed, or `float16` for maximum precision. The RPi 5 handles both extremely well.

```python
# 4. Export to TFLite
model.export(format='tflite', int8=True) 
print("Success! Look for 'best_saved_model' folder for your .tflite file.")
```
